```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
df = pd.read_csv('raw_sales_data.csv')
print(df)
```

```python
print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
print(df.head(10))
```

```python
df.dropna(subset=['Transaction_ID'], inplace=True) # it will drop all rows that have nan values in transaction_id column
```

```python
print(df.isnull().sum())
```

```python
#=============================
#  CLEANING TRANSACTION_IS 
#=============================
```

```python
df.drop_duplicates(subset=['Transaction_ID'], keep='first', inplace=True)
```

```python
print(df)
```

```python
# print(df['Transaction_ID'].duplicated().sum())
```

```python
#=========================
#    CLEANIING PRICES 
#=========================
```

```python
df['Price'] = np.where(df['Price'] == -999.0, np.nan, df['Price']) #np.where(condition, if_true, if_false)
```

```python
mean_price = df['Price'].mean()
```

```python
df['Price'] = df['Price'].fillna(mean_price) # fill nan values with mean of prices
```

```python
df['Price'] = df['Price'].round(2)
```

```python
#=========================
#    CLEANIING QUANTITY 
#=========================
```

```python
df['Quantity'] = np.where((df['Quantity'] < 0) | (df['Quantity'].isnull()), 1.0, df['Quantity'])
```

```python
df['Quantity'] = df['Quantity'].astype(int)
```

```python
print(df['Price'].isnull().sum())
```

```python
print(df['Quantity'].isnull().sum())
```

```python
df
```

```python
'''----- PRODUCT NAME NORMALIZATION -----'''
```

```python
df['Product_Name'] = df['Product_Name'].str.strip().str.title() # it will remove spaces and format to Title 
```

```python
#----- DATE FORMATING -----
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', format='mixed')
```

```python
df['Date'] = df['Date'].ffill() # replace nan with previous valid value
```

```python
#-----CREATING NEW FOLDER Total_Sales
```

```python
df['Total_Sales'] = df['Price'] * df['Quantity']
```

```python
df['Total_Sales'] = df['Total_Sales'].round(2)
```

```python
df
```

```python
# ----- Managing output----
```

```python
product_sales = df.groupby('Product_Name')['Total_Sales'].sum().reset_index()
```

```python
product_sales = product_sales.sort_values(by='Total_Sales', ascending=False)
```

```python
print(product_sales)
```

```python
#======================================
#  LOADING DATA TO SQLITE DATABASE
#======================================
```

```python
conn = sqlite3.connect('product_sales_data.db') # it will create connection with products_sales_data.db, if not exitst it will create this database
```

```python
df.to_sql('Cleaned_sales', conn, if_exists='replace', index=False)
```

```python
product_sales.to_sql('products_summary', conn, if_exists='replace', index=False)
```

```python
conn.close()
```
